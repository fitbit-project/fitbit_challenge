```python
def parse_sleep(records):
    """Parses the sleep data structure."""
    parsed_data = []
    for record_item in records:
        for sleep_log in record_item.get('sleep', []):
            if sleep_log.get('isMainSleep'):
                date_str = sleep_log.get('dateOfSleep')
                minutes = sleep_log.get('minutesAsleep')
                efficiency = sleep_log.get('efficiency')
                if not date_str or minutes is None: continue
                timestamp = datetime.fromisoformat(f"{date_str}T00:00:00")
                parsed_data.append((timestamp, 1, minutes, efficiency))
    return parsed_data
```

```python
all_parsed_data = []
for metric_name in parsers:
    # if metric_name in parsers:
    print(f"Fetching and parsing {metric_name}...")
    parser_func = parsers[metric_name]
    records = device1.get_data(metric_name)
    if records is None:
        print(f"Warning: No data returned for {metric_name}. Skipping.")
        continue
    try:
        parsed_data = parse_sleep(records)
        all_parsed_data.extend(parsed_data)
        print(f"Parsed {len(parsed_data)} points for {metric_name}.")
    except Exception as e:
        print(f"-> Error parsing {metric_name}: {e}")
```

```python
def parse_heart_rate(records):
    """Parses the intraday_heart_rate data structure."""
    parsed_data = []
    for day_data in records:
        date_str = day_data.get('heart_rate_day', [{}])[0].get('activities-heart', [{}])[0].get('dateTime')
        dataset = day_data.get('heart_rate_day', [{}])[0].get('activities-heart-intraday', {}).get('dataset', [])
        if not date_str or not dataset:
            continue
        for point in dataset:
            timestamp = datetime.fromisoformat(f"{date_str}T{point['time']}")
            parsed_data.append((timestamp, 1, 'intraday_heart_rate', point['value']))
    return parsed_data
```

```python
def parse_spo2(records):
    """Parses the intraday_spo2 data structure."""
    parsed_data = []
    for day_data in records:
        # Access the 'dateTime' key for the date of the record
        date_str = day_data.get('dateTime')
        # Access the 'minutes' key, which contains the list of minute-level data
        minutes_data = day_data.get('minutes', [])

        if not date_str or not minutes_data:
            continue

        for point in minutes_data:
            # Access the 'minute' key for the timestamp and 'value' for the SpO2 value
            try:
                timestamp = datetime.datetime.fromisoformat(point['minute'])
                parsed_data.append((timestamp, 1, 'intraday_spo2', point['value']))
            except ValueError as e:
                print(f"Error parsing datetime for minute {point.get('minute')}: {e}")

    return parsed_data

# Test the function with your spo2_rate_records
parsed_spo2_data = parse_spo2(spo2_rate_records)

# Print the first few parsed records to verify
if parsed_spo2_data:
    print("First 10 parsed SpO2 records:")
    for i, record in enumerate(parsed_spo2_data[:10]):
        print(record)
else:
    print("No SpO2 data was parsed.")
```

```python
def parse_breathing_rate(records):
    """Parses the intraday_breath_rate data structure."""
    parsed_data = []
    for day_data in records: # Iterate directly over the list
        br_list = day_data.get('br', []) # Access the list under the 'br' key
        for record in br_list:
            date_str = record.get('dateTime')
            br_data = record.get('value', {})
            if not date_str or not br_data:
                continue

            deep_sleep_summary = br_data.get('deepSleepSummary', {})
            rem_sleep_summary = br_data.get('remSleepSummary', {})
            light_sleep_summary= br_data.get('lightSleepSummary', {})
            full_sleep_summary= br_data.get('fullSleepSummary', {})

            # This metric is a daily summary, not intraday, but we store it with a single timestamp
            try:
                timestamp = datetime.datetime.fromisoformat(f"{date_str}T00:00:00")
                parsed_data.append((timestamp, 1, 'breathing_rate_deep', deep_sleep_summary.get('breathingRate')))
                parsed_data.append((timestamp, 1, 'breathing_rate_rem', rem_sleep_summary.get('breathingRate')))
                parsed_data.append((timestamp, 1, 'breathing_rate_light', light_sleep_summary.get('breathingRate')))
                parsed_data.append((timestamp, 1, 'breathing_rate_full', full_sleep_summary.get('breathingRate')))
            except ValueError as e:
                print(f"Error parsing datetime for date {date_str}: {e}")

    return parsed_data

# Test the function with your breath_rate_records
parsed_br_data = parse_breathing_rate(breath_rate_records)

# Print the first few parsed records to verify
if parsed_br_data:
    print("First 10 parsed Breathing Rate records:")
    for i, record in enumerate(parsed_br_data[:10]):
        print(record)
else:
    print("No Breathing Rate data was parsed.")
```

```python
def parse_activity(records):
    """Parses the intraday_activity data structure."""
    parsed_data = []
    for day_data in records:
        date_str = day_data.get('dateTime')
        value = day_data.get('value')

        if not date_str or value is None:
            continue

        # Assuming activity records are daily summaries, we'll use the start of the day as the timestamp
        try:
            timestamp = datetime.datetime.fromisoformat(f"{date_str}T00:00:00")
            parsed_data.append((timestamp, 1, 'intraday_activity', value))
        except ValueError as e:
            print(f"Error parsing datetime for date {date_str}: {e}")

    return parsed_data

# Test the function with your activity_records
parsed_activity_data = parse_activity(activity_records)

# Print the first few parsed records to verify
if parsed_activity_data:
    print("First 10 parsed Activity records:")
    for i, record in enumerate(parsed_activity_data[:10]):
        print(record)
else:
    print("No Activity data was parsed.")
```

```python
def parse_azm(records):
    """Parses the intraday_active_zone_minute data structure."""
    parsed_data = []
    for day_data in records:
        azm_list = day_data.get('activities-active-zone-minutes-intraday', [])
        for record in azm_list:
            minutes_data = record.get('minutes', [])
            date_str = record.get('dateTime')

            if not date_str or not minutes_data:
                continue

            for point in minutes_data:
                minute_str = point.get('minute')
                value_data = point.get('value', {})

                if not minute_str or not value_data:
                    continue

                try:
                    # Combine the date from the record with the time from the minute data
                    timestamp_str = f"{date_str}T{minute_str}"
                    timestamp = datetime.datetime.fromisoformat(timestamp_str)

                    # Extract individual AZM types and their values
                    fat_burn_azm = value_data.get('fatBurnActiveZoneMinutes')
                    cardio_azm = value_data.get('cardioActiveZoneMinutes')
                    peak_azm = value_data.get('peakActiveZoneMinutes')
                    total_azm = value_data.get('activeZoneMinutes')

                    if fat_burn_azm is not None:
                         parsed_data.append((timestamp, 1, 'azm_fat_burn', fat_burn_azm))
                    if cardio_azm is not None:
                         parsed_data.append((timestamp, 1, 'azm_cardio', cardio_azm))
                    if peak_azm is not None:
                         parsed_data.append((timestamp, 1, 'azm_peak', peak_azm))
                    if total_azm is not None:
                         parsed_data.append((timestamp, 1, 'azm_total', total_azm))

                except ValueError as e:
                    print(f"Error parsing datetime for minute {minute_str}: {e}")

    return parsed_data

# Test the function with your azm_rate_records
parsed_azm_data = parse_azm(azm_rate_records)

# Print the first few parsed records to verify
if parsed_azm_data:
    print("\nFirst 10 parsed AZM records:")
    for i, record in enumerate(parsed_azm_data[:10]):
        print(record)
else:
    print("\nNo AZM data was parsed.")
```

```python
def parse_hrv(records):
    """Parses the intraday_hrv data structure."""
    parsed_data = []
    for day_data in records:
        hrv_list = day_data.get('hrv', [])
        for record in hrv_list:
            minutes_data = record.get('minutes', [])

            if not minutes_data:
                continue

            for point in minutes_data:
                minute_str = point.get('minute')
                value_data = point.get('value', {})

                if not minute_str or not value_data:
                    continue

                try:
                    # The minute string contains the full timestamp
                    timestamp = datetime.datetime.fromisoformat(minute_str)

                    # Extract individual HRV metrics
                    rmssd = value_data.get('rmssd')
                    coverage = value_data.get('coverage')
                    hf = value_data.get('hf')
                    lf = value_data.get('lf')

                    if rmssd is not None:
                        parsed_data.append((timestamp, 1, 'hrv_rmssd', rmssd))
                    if coverage is not None:
                        parsed_data.append((timestamp, 1, 'hrv_coverage', coverage))
                    if hf is not None:
                        parsed_data.append((timestamp, 1, 'hrv_hf', hf))
                    if lf is not None:
                        parsed_data.append((timestamp, 1, 'hrv_lf', lf))

                except ValueError as e:
                    print(f"Error parsing datetime for minute {minute_str}: {e}")

    return parsed_data

# Test the function with your hrv_rate_records
parsed_hrv_data = parse_hrv(hrv_rate_records)

# Print the first few parsed records to verify
if parsed_hrv_data:
    print("\nFirst 10 parsed HRV records:")
    for i, record in enumerate(parsed_hrv_data[:10]):
        print(record)
else:
    print("\nNo HRV data was parsed.")
```

```python
def streaming_keys(path):
    """
    Iterate over the file once, never loading the whole thing.
    Returns a set of every dotted path key seen.
    """
    keys = set()
    with open(path, "rb") as f:            # ijson needs bytes
        for prefix, event, value in ijson.parse(f):
            # prefix is like 'users.item.address.city'
            if event in {"map_key", "string", "number", "boolean", "null"}:
                # split prefix on '.' then rebuild dotted keys for every ancestor
                parts = prefix.split(".")
                for i in range(1, len(parts) + 1):
                    keys.add(".".join(parts[:i]))
    return keys

if __name__ == "__main__":
    keys = streaming_keys("synthetic_fitbit_data.json")
    for k in sorted(keys):
        print(k)
```
